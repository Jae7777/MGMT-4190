```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/lmanikon/lmanikon.github.io/refs/heads/master/teaching/datasets/nba.csv')
df
```

```python
df = df.dropna(how='all')
df
```

```python
df['College'] = df['College'].fillna('No College')
df
```

```python
# fill salary with mean of salary from team
gby = df.groupby('Team')
for team, subdf in gby:
  print(team, len(subdf), round(subdf['Salary'].mean(), 2))
```

```python
# gorupby based on the position
gby = df.groupby('Position')
for pos, subdf in gby:
  print(pos, len(subdf), round(subdf['Salary'].mean(), 2))
```
