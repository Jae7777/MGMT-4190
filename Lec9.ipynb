### Data preprocessing

```python
#1. Lets read the file to a dataframe and print all the rows and columns
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/lmanikon/lmanikon.github.io/master/teaching/datasets/airlineSatisfaction.csv')
df.head()
```

```python
#FInding the unique  values of satisfaction column
print(df['satisfaction'].unique())
```

```python
#2. Find the total number of data points in this dataset?
print(df.shape[0])
```

```python
#3. Total number of attributes in this dataset
print(df.shape[1])
```

```python
#4. Check if there are any missing values in this dataset -- if yes, print true or else print false
print(df.isnull().any().any())
```

```python
#5. If there are missing values, print the total number of missing values
print(df.isnull().sum().sum())
```

```python
#6. Now drop all the rows with atleast one missing value
df = df.dropna(how="any")
print(df.shape)
print(df.isnull().any().sum())
```

```python
#7. Lets print the summary statistics
print(df.describe())
```

```python
#8. Lets say drop the column 'id'
df = df.drop(columns=['id'])
print(df)

```

```python
#9. Categorical attributes to numerical attributes

#Gender
#Male -- 1; Female -- 0



#Can you write a Lambda function?
df['Gender'] = df['Gender'].apply(lambda x: 1 if x == 'Male' else 0)
print(df)

```

```python
#10. Take the 'satisfaction' column and perform this conversion -- "neutral or dissatisfied": -1; "satisfied":1

df['satisfaction'] = df['satisfaction'].apply(lambda x: -1 if x == 'neutral or dissatisfied' else 1)
print(df)
```

```python
#11. Flight distance -- Binarization

#FD >=1000 -- 1
#Otherwise, -- 0
df['Flight Distance'] = df['Flight Distance'].apply(lambda x: 1 if x >= 1000 else 0)
print(df)

```

```python
#12. Age attribute -- Discretize
#[7,27) -- 1
#[27,51) -- 2
#[51,85) -- 3
df['Age'] = df['Age'].apply(
  lambda x: 1 if x >= 7 and x < 27 else 2 if x >= 27 and x < 51 else 3 if x >= 51 and x < 85 else x
)

print(df)
```

```python
#13. perform random sampling
#Sample 100 random datapoints from our dataframe df -- sample without replacement
print(df.sample(n=100, replace=False))

```

```python
#14. Split the dataframe df into X and y -- X-- corresponds to feature set; y -- class label

#X - excluding 'satisfaction'
#y - 'satisfaction'
#y = f(X) -- y-class label variable; X -- feature set
X = df.drop(columns=['satisfaction'])
y = df['satisfaction']
print(X)  
print(y)

```

```python
#15. Aggregate -- based on the type of travel -- summing up all the values
print(df.groupby('Type of Travel').sum())
```
