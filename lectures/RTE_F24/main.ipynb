```python
import pandas as pd
import numpy as np
import plotly.express as px

pd.set_option('display.max_columns', None)

df = pd.read_csv('BathSoapHousehold.csv')
df.head()
```

```python
group_by_CHILDREN = df.groupby('CHILD')
group_by_CHILDREN.describe()
fig = px.histogram(df, x='CHILD', y=df['Total Volume'], color='CHILD', histfunc='avg')
fig.show()

fig = px.histogram(df, x='CHILD', y=df['Vol/Tran'], color='CHILD', histfunc='avg')
fig.show()

px.histogram(df, x='CHILD', y=df['Avg. Price '], color='CHILD', histfunc='avg').show()
```

```python
import seaborn as sns
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_std = scaler.fit_transform(df)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=30, random_state=0).fit(df_std)
df['cluster'] = kmeans.labels_

fig = px.scatter_3d(df, x='CHILD', y='Total Volume', z='Vol/Tran', color='cluster')
fig.show()
```
